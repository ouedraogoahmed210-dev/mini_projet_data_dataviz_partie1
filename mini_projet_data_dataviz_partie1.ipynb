# PHASE A — Data Process

## A.1 — Import des données


````python
import pandas as pd

URL = "https://raw.githubusercontent.com/davidmegginson/ourairports-data/main/airports.csv"

# TODO : importer le CSV avec pandas
df = pd.read_csv(URL)
print(df.shape)
df.head()
````

## A.2 — Prétraitement

- Ne garder que les colonnes utiles : `ident`, `name`, `type`, `iso_country`, `latitude_deg`, `longitude_deg`, `elevation_ft`, `scheduled_service`.
- Renommer respectivement en : `ID`, `airport_name`, `type`, `country`, `lat`, `long`, `elevation`, `service`.


````python
# TODO : ne garder que les colonnes utiles :
#        ident, name, type, iso_country, latitude_deg, longitude_deg, elevation_ft, scheduled_service
df = df[["ident", "name", "type", "iso_country", "latitude_deg", "longitude_deg", "elevation_ft", "scheduled_service"]]

# TODO : renommer ces colonnes respectivement en :
#        ID, airport_name, type, country, lat, long, elevation, service
df = df.rename(columns={
    "ident": "ID",
    "name": "airport_name",
    "type": "type",
    "iso_country": "country",
    "latitude_deg": "lat",
    "longitude_deg": "long",
    "elevation_ft": "elevation",
    "scheduled_service": "service"
})

df.head()
````

## A.3 — Opérations individualisées

Toutes les opérations suivantes portent uniquement sur la **région qui vous a été assignée à l'Étape 0**.


````python
# TODO : filtrer le DataFrame sur les pays de votre région assignée
#        (indice : méthode .isin())
df_region = df[df["country"].isin(['PL', 'CZ', 'SK', 'HU', 'RO', 'BG', 'UA'])]  # votre code ici

print(f"Nombre d'aéroports dans la région {region_name} : {len(df_region)}")
df_region.head()
````

````python
# TODO : extraire les aéroports dont le nom contient un mot-clé de votre choix (ex. 'International')
df_international = df_region[df_region["airport_name"].str.contains("International", case=False, na=False)]
# Sauvegardez le résultat dans international.csv
df_international.to_csv("international.csv", index=False)

# TODO : extraire les aéroports dont le nom commence entre deux lettres de votre choix (ex. A-M)
df_range_alpha = df_region[df_region["airport_name"].str.upper().str[0].between("A", "M")]
# Sauvegardez le résultat dans range_alpha.csv
df_range_alpha.to_csv("range_alpha.csv", index=False)

# TODO : identifiez (avec nsmallest/nlargest) l'aéroport à l'altitude minimale et maximale de la région
#        (nom, pays, altitude)
colonnes_affichage = ["airport_name", "country", "elevation"]
aeroport_max = df_region.nlargest(1, "elevation")[colonnes_affichage]
aeroport_min = df_region.nsmallest(1, "elevation")[colonnes_affichage]

print("\nAéroport à l'altitude MAXIMALE ")
print(aeroport_max.to_string(index=False))

print("\nAéroport à l'altitude MINIMALE")
print(aeroport_min.to_string(index=False))
````

### Synthèse par pays

Pour chaque pays de votre région assignée, calculez :
- le nombre total d'aéroports,
- le nombre de `large_airport`,
- le nombre d'aéroports avec service régulier (`service == "yes"`).


````python
# TODO : pour chaque pays, calculez :
#        - le nombre total d'aéroports
#        - le nombre de 'large_airport'
#        - le nombre d'aéroports avec service régulier (service == 'yes')
# Indice : df_region.groupby("country").agg(...)
synthese = df_region.groupby("country", as_index=False).agg(
    nombre_total_airport=("ID", "count"),
    nombre_large_airport=("type", lambda x: (x == "large_airport").sum()),
    service_regulier=("service", lambda x: (x == "yes").sum())
)

synthese
````

````python
# TODO : identifier les pays SANS AUCUN grand aéroport en service régulier
#        (type == 'large_airport' ET service == 'yes')
df_large_actif = df_region[(df_region["type"] == "large_airport") & (df_region["service"] == "yes")]  # votre code ici
pays_sous_desservis = set(synthese["country"]) - set(df_large_actif["country"])   # votre code ici (différence d'ensembles)

print("Pays sous-desservis (aucun grand aéroport actif) :", pays_sous_desservis)

synthese.to_csv("synthese_pays.csv", index=False)
````

## B.4 — Avant : graphique par défaut

Produisez d'abord un graphique **sans aucune amélioration**, tel que Matplotlib le génère par défaut, sur la variable de la table `synthese` qui illustre votre message (ex. `nb_service_regulier`, `nb_large`...).


````python
import matplotlib.pyplot as plt

# TODO : choisissez la colonne de 'synthese' qui illustre votre message
colonne_a_visualiser = "service_regulier"

fig, ax = plt.subplots()
ax.bar(synthese["country"], synthese[colonne_a_visualiser])
ax.set_title(f"{colonne_a_visualiser} by country")
plt.show()
````

## B.5 — Choix du graphique (justification)

Le graphique recommandé est le graphique en barres horizontales (ou Bar Chart horizontal). Il permet de lire facilement les noms des pays de gauche à droite et de comparer instantanément la longueur des barres.

Pourquoi un pie chart, un donut chart ou un graphique 3D seraient de mauvais choix ici ? 

Pourquoi un pie chart, un donut chart?
Difficulté de lecture : Notre œil compare très mal les angles et les surfaces de "parts de gâteau" par rapport à des barres alignées.
Trop de catégories : Avec 7 pays, le graphique devient immédiatement surchargé et illisible.

Pourquoi la 3D est un mauvais choix ?

Déformation visuelle : L'effet de perspective fait paraître les éléments au premier plan plus grands qu'à l'arrière-plan.

 Lecture faussée : Cette distorsion empêche de lire les valeurs exactes et masque les vrais écarts entre les pays.


## B.6 — Decluttering pas à pas

Reproduisez la méthode vue en cours, une étape à la fois. Pour chaque étape, le code de l'étape précédente est repris et complété.


````python
# Étape 1 — Passage en barres horizontales (recommandé pour les catégories, cf. cours)
#           + suppression de la bordure du graphique

synthese_sorted = synthese.sort_values(colonne_a_visualiser)

fig, ax = plt.subplots()
# TODO : tracez un graphique en barres horizontales avec synthese_sorted (ax.barh)
ax.barh(synthese_sorted["country"], synthese_sorted[colonne_a_visualiser])
# TODO : supprimez les 4 bordures (spines) du graphique
#        indice : ax.spines["..."].set_visible(False)
for spine in ["top", "right", "left", "bottom"]:
    ax.spines[spine].set_visible(False)

plt.show()
````

````python
# Étape 2 — Nettoyer les labels d'axes (supprimer ce qui n'apporte rien)

fig, ax = plt.subplots()
# TODO : reprenez le code de l'étape 1 (barh + suppression des bordures)
ax.barh(synthese_sorted["country"], synthese_sorted[colonne_a_visualiser])

for spine in ["top", "right", "left", "bottom"]:
    ax.spines[spine].set_visible(False)
# TODO : supprimez les graduations de l'axe x et son label
#        indice : ax.tick_params(axis="x", length=0), ax.set_xticklabels([]), ax.set_xlabel("")
ax.tick_params(axis="x", length=0)
ax.set_xticklabels([])
ax.set_xlabel("")
plt.show()
````

````python
# Étape 3 — Labellisation directe des données + couleur intentionnelle
#           (mettre en évidence UN SEUL pays — celui de votre message — le reste en gris)

# TODO : remplacez par le pays que vous voulez mettre en évidence (cohérent avec votre storyboard)
pays_a_mettre_en_evidence ="RO"  

# TODO : construisez une liste de couleurs : une couleur vive pour le pays mis en évidence,
#        gris pour tous les autres (ex. via une compréhension de liste sur synthese_sorted["country"])
couleurs = ["#1f77b4" if c in pays_a_mettre_en_evidence else "#d3d3d3" for c in synthese_sorted["country"]]

fig, ax = plt.subplots()
# TODO : reprenez le code des étapes 1 et 2, en utilisant 'couleurs' au lieu d'une couleur unique
bars = ax.barh(synthese_sorted["country"], synthese_sorted[colonne_a_visualiser], color=couleurs)

for spine in ["top", "right", "left", "bottom"]:
    ax.spines[spine].set_visible(False)

ax.tick_params(axis="x", length=0)
ax.set_xticklabels([])
ax.set_xlabel("")

# TODO : ajoutez la valeur directement au bout de chaque barre (label direct, pas de légende)
#        indice : ax.text(...) pour chaque barre
for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.3,                              
            bar.get_y() + bar.get_height()/2,          
            f'{int(width)}',                           
            ha='left', va='center', fontsize=10,
            fontweight="bold" if width == synthese_sorted[colonne_a_visualiser].max() else "normal")
plt.show()
````

````python
# Étape 4 — Version finale : titre déclaratif (l'insight, pas une description)
#           Reprend tout le code des étapes précédentes.

# TODO : remplacez par votre titre déclaratif (l'insight, formulé comme une phrase complète)
titre_declaratif = "Fracture aérienne : La Roumanie en tête, quatre fois mieux desservie que la Slovaquie"

fig, ax = plt.subplots(figsize=(8, 5))

# TODO : reprenez l'intégralité du code des étapes 1 à 3 (barh, couleurs, suppression des
#        bordures et axes, labels directs)
bars = ax.barh(synthese_sorted["country"], synthese_sorted[colonne_a_visualiser], color=couleurs)

# Suppression des 4 bordures (spines) - Étape 1
for spine in ["top", "right", "left", "bottom"]:
    ax.spines[spine].set_visible(False)

# Nettoyage des axes et graduations du bas - Étape 2
ax.tick_params(axis="x", length=0)
ax.set_xticklabels([])
ax.set_xlabel("")
ax.tick_params(axis="y", length=0)

# Ajout des labels directs au bout de chaque barre - Étape 3
for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.3, 
            bar.get_y() + bar.get_height()/2, 
            f'{int(width)}', 
            ha='left', va='center', fontsize=10, 
            fontweight="bold" if width == synthese_sorted[colonne_a_visualiser].max() else "normal")

# Titre déclaratif portant l'insight
ax.set_title(titre_declaratif, fontsize=13, fontweight="bold", loc="left", pad=15)

plt.tight_layout()

# Sauvegarde automatique du graphique final exigé
plt.savefig("graphique_final.png", dpi=150, bbox_inches="tight")
plt.show()
````
